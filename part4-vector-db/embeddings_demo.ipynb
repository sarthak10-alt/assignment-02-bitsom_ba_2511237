```python
# First, we install the library that does the 'thinking'
# !pip install sentence-transformers

from sentence_transformers import SentenceTransformer, util

# 1. We pick a 'Brain' (Model)
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Your 10 sentences across Cricket, Cooking, and Cybersecurity
sentences = [
    "The batsman hit a massive six over long-on.",
    "Fast bowlers use swing to deceive the batter.",
    "A century in a test match requires great patience.",
    "Whisk the eggs until they are fluffy and light.",
    "Slow-cook the stew for four hours for best flavor.",
    "Add a pinch of salt to balance the sweetness.",
    "Always use a strong password with special characters.",
    "Two-factor authentication adds a layer of security.",
    "Phishing emails often try to steal your login info.",
    "Firewalls block unauthorized access to the network."
]

# 3. Turn sentences into 'Vectors' (Math numbers)
embeddings = model.encode(sentences)

# 4. Find the most similar sentences to a new query
query = "The bowler took three wickets in one over"
query_embedding = model.encode(query)

# Compute similarities
hits = util.semantic_search(query_embedding, embeddings, top_k=2)

print(f"Query: {query}")
for hit in hits[0]:
    print(f"Match: {sentences[hit['corpus_id']]} (Score: {hit['score']:.4f})")
```
